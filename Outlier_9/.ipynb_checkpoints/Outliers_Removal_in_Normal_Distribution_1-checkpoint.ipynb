```python
# Approach 2

# calculate the Z-Score
df['cgpa_Z-Score'] = (df['cgpa'] - df['cgpa'].mean()) / df['cgpa'].std()
```

```python
df.head()
```

```python
df[df['cgpa_Z-Score'] > 3]
```

```python
df[df['cgpa_Z-Score'] < -3]
```

```python
df[(df['cgpa_Z-Score'] > 3) | (df['cgpa_Z-Score'] < -3)]
```

```python
# Trimming

new_df2 = df[(df['cgpa_Z-Score'] < 3) & (df['cgpa_Z-Score'] > -3)]

print(new_df2.shape)
new_df2
```

```python
plt.figure(figsize = (16,5))
plt.subplot(1,2,1)
plt.title("Old_df")
sns.distplot(df['cgpa'])

plt.subplot(1,2,2)
plt.title("New_df2")
sns.distplot(new_df2['cgpa'])



plt.show()
```

# Capping

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
df['cgpa'] = np.where(
    df['cgpa'] > upper_limit,
    upper_limit,
    np.where(
        df['cgpa'] < lower_limit,
        lower_limit,
        df['cgpa']
    )
)
```
