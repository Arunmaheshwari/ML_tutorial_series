# Working with Dates

```python
# Converting to datatime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
date['date_year'] = date['date'].dt.year
date.head()
```

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
date['date_dow'] = date['date'].dt.dayofweek

# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.head()
```

```python
date.drop(columns = ['product_id', 'city_id', 'orders']).head()
```

```python
# is weekend

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Monday']), 1,0)
```

```python
date.head()
```

```python
# Extract week of the year
date['date_week'] = date['date'].dt.isocalendar().week

date.head()
```

```python
date['date_quarter'] = date['date'].dt.quarter
#date.drop(columns = ['product_id', 'city_id', 'orders'], inplace = True)
date.head()
```

```python
date['semester'] = np.where(date['date_quarter'].isin([1,2]), 1,2)
date.head()
```

```python
# Extract Time elapsed between dates
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```
