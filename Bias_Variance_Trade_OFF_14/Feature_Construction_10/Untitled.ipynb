# Applying feature construction

```python
X['Family'] = X['SibSp'] + X['Parch'] +1
```

```python
X.head()
```

```python
def myfuc(num):
    if num == 1:
        return 0
    elif num >1 and num <=4:
        return 1
    else:
        return 2
```

```python
myfuc(4)
```

```python
myfuc(5)
```

```python
myfuc(1)
```

```python
X['Family_type'] = X['Family'].apply(myfuc)
```

```python
X.head()
```

```python
X.drop(columns = ['Parch', 'SibSp', 'Family'], inplace = True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), X,y, scoring = 'accuracy', cv = 20))
```
